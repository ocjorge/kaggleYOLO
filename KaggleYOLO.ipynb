```python
# ==============================================================================
# PASO 0: INSTALACIÃ“N DE LIBRERÃAS NECESARIAS
# ==============================================================================
print("Paso 0: Instalando librerÃ­as...")
!pip install ultralytics -q
!pip install pylabel -q
!pip install pyyaml -q

import os
import yaml
import numpy as np
import matplotlib.pyplot as plt
import cv2
import glob
import zipfile # Para descomprimir
from ultralytics import YOLO
import ultralytics

print("LibrerÃ­as instaladas y mÃ³dulos importados.")
print(f"Ultralytics versiÃ³n: {ultralytics.__version__}")

# ==============================================================================
# PASO 0.5: DESCOMPRIMIR EL DATASET
# ==============================================================================
print("\nPaso 0.5: Descomprimiendo el dataset...")

# --------- MODIFICA ESTAS VARIABLES SI ES NECESARIO ---------
ZIP_FILE_NAME = "vehicle_dataset.zip"  # <<<<<<< CAMBIA ESTO al nombre de tu archivo ZIP
# Asume que el archivo ZIP estÃ¡ en el directorio raÃ­z de Colab (/content/)
COLAB_BASE_PATH = "/content/"
ZIP_FILE_PATH = os.path.join(COLAB_BASE_PATH, ZIP_FILE_NAME)

# Directorio donde se extraerÃ¡ el contenido del ZIP
EXTRACT_DIR_NAME = "datasets_extracted"
EXTRACT_TO_DIR = os.path.join(COLAB_BASE_PATH, EXTRACT_DIR_NAME)

# Nombre de la carpeta principal DENTRO del ZIP que contiene 'train', 'valid', etc.
# Si tu ZIP extrae directamente 'train/', 'valid/', etc., deja esto como "" o ajusta BASE_DATASET_PATH despuÃ©s.
# Ejemplo: si tu ZIP contiene "vehicle dataset/train", "vehicle dataset/valid", entonces pon "vehicle dataset"
# Si tu ZIP contiene directamente "train", "valid", entonces pon "" y BASE_DATASET_PATH serÃ¡ EXTRACT_TO_DIR
EXPECTED_TOP_LEVEL_FOLDER_INSIDE_ZIP = "vehicle dataset" # <<<<<<< AJUSTA ESTO
# -----------------------------------------------------------

if os.path.exists(ZIP_FILE_PATH):
    print(f"Archivo ZIP encontrado: {ZIP_FILE_PATH}")
    print(f"Extrayendo a: {EXTRACT_TO_DIR}...")

    # Crear el directorio de extracciÃ³n si no existe
    os.makedirs(EXTRACT_TO_DIR, exist_ok=True)

    # Descomprimir usando el comando unzip (mÃ¡s robusto para estructuras complejas)
    # El -o es para sobrescribir sin preguntar, -q para modo silencioso
    !unzip -q -o "{ZIP_FILE_PATH}" -d "{EXTRACT_TO_DIR}"

    print("DescompresiÃ³n completada.")

    # Actualizar BASE_DATASET_PATH para que apunte a la carpeta extraÃ­da correcta
    if EXPECTED_TOP_LEVEL_FOLDER_INSIDE_ZIP:
        BASE_DATASET_PATH = os.path.join(EXTRACT_TO_DIR, EXPECTED_TOP_LEVEL_FOLDER_INSIDE_ZIP)
    else:
        # Si no hay una carpeta de nivel superior dentro del zip, los directorios train/valid estÃ¡n directamente en EXTRACT_TO_DIR
        BASE_DATASET_PATH = EXTRACT_TO_DIR

    print(f"El dataset base ahora estÃ¡ en: {BASE_DATASET_PATH}")

    # Verificar si la ruta base del dataset extraÃ­do existe
    if not os.path.exists(BASE_DATASET_PATH):
        print(f"ERROR: La ruta base del dataset extraÃ­do {BASE_DATASET_PATH} no existe.")
        print(f"Verifica la variable 'EXPECTED_TOP_LEVEL_FOLDER_INSIDE_ZIP' y la estructura de tu ZIP.")
        print(f"Contenido de {EXTRACT_TO_DIR}: {os.listdir(EXTRACT_TO_DIR)}")
        # Detener la ejecuciÃ³n si no se encuentra el dataset
        raise FileNotFoundError(f"Dataset no encontrado en {BASE_DATASET_PATH} despuÃ©s de la extracciÃ³n.")
else:
    print(f"ERROR: Archivo ZIP '{ZIP_FILE_PATH}' no encontrado.")
    print("Por favor, sube tu archivo ZIP al directorio raÃ­z de Colab ('/content/') y asegÃºrate de que 'ZIP_FILE_NAME' sea correcto.")
    # Detener la ejecuciÃ³n si no se encuentra el ZIP
    raise FileNotFoundError(f"Archivo ZIP {ZIP_FILE_PATH} no encontrado.")

# ==============================================================================
# PASO 1: DEFINICIÃ“N DE RUTAS Y VARIABLES PRINCIPALES (USANDO DATOS EXTRAÃDOS)
# ==============================================================================
print("\nPaso 1: Definiendo rutas y variables...")

TRAIN_DIR = os.path.join(BASE_DATASET_PATH, 'train')
VALID_DIR = os.path.join(BASE_DATASET_PATH, 'valid') # AsegÃºrate que 'valid' existe y tiene images/labels

CLASS_NAMES = ['car', 'threewheel', 'bus', 'truck', 'motorbike', 'van']
NUM_CLASSES = len(CLASS_NAMES)

# Ruta donde guardaremos el archivo data.yaml (en el entorno de trabajo de Colab)
# /kaggle/working/ es persistente entre sesiones de ejecuciÃ³n de Colab (hasta que se reinicia el runtime)
# /content/ no es persistente.
# Usaremos /kaggle/working/ para el YAML y los resultados del entrenamiento
WORKING_DIR = "/kaggle/working/"
os.makedirs(WORKING_DIR, exist_ok=True) # Asegurarse de que exista
DATA_YAML_PATH = os.path.join(WORKING_DIR, 'data.yaml')


print(f"Directorio de entrenamiento: {TRAIN_DIR}")
print(f"Directorio de validaciÃ³n: {VALID_DIR}")
print(f"Clases: {CLASS_NAMES}")
print(f"NÃºmero de clases: {NUM_CLASSES}")
print(f"Archivo YAML se guardarÃ¡ en: {DATA_YAML_PATH}")

if not os.path.exists(TRAIN_DIR) or not os.path.exists(os.path.join(TRAIN_DIR, 'images')) or not os.path.exists(os.path.join(TRAIN_DIR, 'labels')):
    print(f"ADVERTENCIA: El directorio de entrenamiento {TRAIN_DIR} o sus subcarpetas 'images'/'labels' no existen o estÃ¡n mal configurados.")
    print(f"Contenido de {BASE_DATASET_PATH}: {os.listdir(BASE_DATASET_PATH) if os.path.exists(BASE_DATASET_PATH) else 'No existe'}")
    if os.path.exists(TRAIN_DIR):
        print(f"Contenido de {TRAIN_DIR}: {os.listdir(TRAIN_DIR)}")

if not os.path.exists(VALID_DIR) or not os.path.exists(os.path.join(VALID_DIR, 'images')) or not os.path.exists(os.path.join(VALID_DIR, 'labels')):
    print(f"ADVERTENCIA: El directorio de validaciÃ³n {VALID_DIR} o sus subcarpetas 'images'/'labels' no existen o estÃ¡n mal configurados.")
    print("Si no tienes un conjunto de validaciÃ³n, el entrenamiento podrÃ­a fallar o no ser Ã³ptimo.")
    if os.path.exists(VALID_DIR):
        print(f"Contenido de {VALID_DIR}: {os.listdir(VALID_DIR)}")


# ==============================================================================
# PASO 2: EXPLORACIÃ“N DE DATOS (OPCIONAL PERO RECOMENDADO)
# ==============================================================================
print("\nPaso 2: ExploraciÃ³n de datos...")

def image_plot(file_path):
    img = cv2.imread(file_path)
    if img is None:
        print(f"Error: No se pudo cargar la imagen {file_path}")
        return None
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    plt.imshow(img_rgb)
    plt.title(os.path.basename(file_path))
    plt.axis('off')
    plt.show()
    return img_rgb

# Ejemplo de visualizaciÃ³n de una imagen del conjunto de entrenamiento
# AsegÃºrate que la ruta sea correcta y la imagen exista
# Vamos a intentar listar alguna imagen si la especÃ­fica no se encuentra
sample_image_to_find = '00043_GMC Savana Van 2012.jpg' # Ejemplo que tenÃ­as
train_images_path = os.path.join(TRAIN_DIR, 'images')

if os.path.exists(os.path.join(train_images_path, sample_image_to_find)):
    sample_image_path = os.path.join(train_images_path, sample_image_to_find)
    print(f"Mostrando imagen de ejemplo: {sample_image_path}")
    image_plot(sample_image_path)
else:
    print(f"Imagen de ejemplo '{sample_image_to_find}' no encontrada en: {train_images_path}")
    try:
        sample_images_list = glob.glob(os.path.join(train_images_path, '*.jpg')) + glob.glob(os.path.join(train_images_path, '*.png'))
        if sample_images_list:
            print(f"Mostrando otra imagen de ejemplo: {sample_images_list[0]}")
            image_plot(sample_images_list[0])
        else:
            print(f"No se encontraron imÃ¡genes (.jpg o .png) en {train_images_path}")
    except Exception as e:
        print(f"Error listando imÃ¡genes: {e}")


def count_labels_in_files(label_files_dir, num_classes):
    counts = [0] * num_classes
    if not os.path.exists(label_files_dir):
        print(f"Directorio de etiquetas no encontrado: {label_files_dir}")
        return counts

    label_files = glob.glob(os.path.join(label_files_dir, '*.txt'))
    if not label_files:
        print(f"No se encontraron archivos .txt en {label_files_dir}")
        return counts

    for file_path in label_files:
        try:
            with open(file_path, 'r') as file:
                for line in file:
                    parts = line.split()
                    if parts:
                        class_id = int(parts[0])
                        if 0 <= class_id < num_classes:
                            counts[class_id] += 1
                        else:
                            print(f"ID de clase fuera de rango ({class_id}) en archivo: {file_path}")
        except Exception as e:
            print(f"Error procesando archivo {file_path}: {e}")
    return counts

train_labels_dir = os.path.join(TRAIN_DIR, 'labels')
print(f"\nContando etiquetas en el conjunto de entrenamiento ({train_labels_dir})...")
label_counts = count_labels_in_files(train_labels_dir, NUM_CLASSES)

print("\nDistribuciÃ³n de clases en el conjunto de entrenamiento:")
vehicle_counts = {}
for i, class_name in enumerate(CLASS_NAMES):
    vehicle_counts[class_name] = label_counts[i]
    print(f"{class_name}: {label_counts[i]}")

if sum(vehicle_counts.values()) > 0 : # Solo graficar si hay datos
    plt.figure(figsize=(10, 5))
    plt.bar(vehicle_counts.keys(), vehicle_counts.values(), color='skyblue')
    plt.xlabel('Tipo de vehÃ­culo')
    plt.ylabel('NÃºmero de instancias')
    plt.title('DistribuciÃ³n de clases en el conjunto de entrenamiento')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron etiquetas para graficar la distribuciÃ³n.")

# ==============================================================================
# PASO 3: CREACIÃ“N DEL ARCHIVO data.yaml
# ==============================================================================
print("\nPaso 3: Creando el archivo data.yaml...")

data_yaml_content = {
    'path': BASE_DATASET_PATH, # Ruta absoluta al directorio base del dataset
    'train': 'train/images',   # Ruta relativa a 'path' para imÃ¡genes de entrenamiento
    'val': 'valid/images',     # Ruta relativa a 'path' para imÃ¡genes de validaciÃ³n
    # 'test': 'test/images',  # Descomentar si tienes un conjunto de test
    'nc': NUM_CLASSES,
    'names': CLASS_NAMES
}
# NOTA: YOLOv8 espera que los archivos de etiquetas (.txt) estÃ©n en directorios 'labels'
# paralelos a los directorios 'images'. Ejemplo:
# path/train/images/
# path/train/labels/
# path/valid/images/
# path/valid/labels/
# El archivo YAML no necesita especificar los directorios 'labels' si siguen esta convenciÃ³n.

try:
    with open(DATA_YAML_PATH, 'w') as f:
        yaml.dump(data_yaml_content, f, sort_keys=False, Dumper=yaml.SafeDumper) # Usar SafeDumper
    print(f"Archivo data.yaml creado exitosamente en {DATA_YAML_PATH}")
    print("\nContenido de data.yaml:")
    with open(DATA_YAML_PATH, 'r') as f:
        print(f.read())
except Exception as e:
    print(f"Error al crear o escribir el archivo data.yaml: {e}")

# ==============================================================================
# PASO 4: ENTRENAMIENTO DEL MODELO YOLOv8
# ==============================================================================
print("\nPaso 4: Entrenando el modelo YOLOv8...")

print("\nRealizando chequeo de Ultralytics...")
try:
    ultralytics.checks()
except Exception as e:
    print(f"Error durante ultralytics.checks(): {e}")

model = YOLO("yolov8n.pt")
print(f"\nModelo YOLO('yolov8n.pt') cargado.")

BATCH_SIZE = 16
EPOCHS = 20
IMG_SIZE = 640

print(f"Iniciando entrenamiento con batch_size={BATCH_SIZE}, epochs={EPOCHS}, imgsz={IMG_SIZE}")
print(f"Usando data YAML: {DATA_YAML_PATH}")

if not os.path.exists(DATA_YAML_PATH):
    print(f"ERROR CRÃTICO: El archivo {DATA_YAML_PATH} no existe. No se puede entrenar.")
else:
    try:
        results = model.train(
            data=DATA_YAML_PATH,
            batch=BATCH_SIZE,
            epochs=EPOCHS,
            imgsz=IMG_SIZE,
            project=os.path.join(WORKING_DIR, 'runs/detect'),
            name='train_experiment_unzip'
        )
        print("Entrenamiento completado.")
        LATEST_RUN_DIR = os.path.join(WORKING_DIR, 'runs/detect/train_experiment_unzip')
        # Si se crea train_experiment_unzip2, train_experiment_unzip3, etc.
        # buscar el mÃ¡s reciente
        potential_dirs = glob.glob(os.path.join(WORKING_DIR, 'runs/detect/train_experiment_unzip*'))
        if potential_dirs:
            LATEST_RUN_DIR = max(potential_dirs, key=os.path.getctime)

        print(f"Resultados del entrenamiento guardados en: {LATEST_RUN_DIR}")
        BEST_MODEL_PATH_TRAIN = os.path.join(LATEST_RUN_DIR, 'weights/best.pt')
        print(f"El mejor modelo se guardÃ³ en: {BEST_MODEL_PATH_TRAIN}")
    except Exception as e:
        print(f"OcurriÃ³ un error durante el entrenamiento: {e}")
        import traceback
        traceback.print_exc()

# ==============================================================================
# PASO 5: EVALUACIÃ“N E INFERENCIA (USANDO EL MODELO ENTRENADO)
# ==============================================================================
print("\nPaso 5: EvaluaciÃ³n e inferencia...")

BEST_MODEL_PATH = None
try:
    # Buscar la carpeta de entrenamiento mÃ¡s reciente
    list_of_run_dirs = glob.glob(os.path.join(WORKING_DIR, 'runs/detect/train_experiment_unzip*'))
    if list_of_run_dirs:
        latest_run_dir = max(list_of_run_dirs, key=os.path.getctime)
        BEST_MODEL_PATH = os.path.join(latest_run_dir, 'weights/best.pt')
        print(f"Usando el modelo entrenado: {BEST_MODEL_PATH}")

    if BEST_MODEL_PATH and os.path.exists(BEST_MODEL_PATH):
        trained_model = YOLO(BEST_MODEL_PATH)

        print("\nEvaluando el modelo en el conjunto de validaciÃ³n...")
        try:
            metrics = trained_model.val(data=DATA_YAML_PATH) # Especificar data para asegurar que usa el conjunto correcto
            if metrics and hasattr(metrics, 'box') and hasattr(metrics.box, 'map50'):
                 print("MÃ©tricas de validaciÃ³n (mAP50):", metrics.box.map50)
            else:
                print("No se pudieron obtener las mÃ©tricas de validaciÃ³n detalladas.")
        except Exception as e:
            print(f"Error durante la validaciÃ³n: {e}")

        print("\nRealizando inferencia en una imagen de ejemplo...")
        valid_images_path = os.path.join(VALID_DIR, 'images')
        if os.path.exists(valid_images_path):
            valid_images = glob.glob(os.path.join(valid_images_path, '*.jpg')) + glob.glob(os.path.join(valid_images_path, '*.png'))
            if valid_images:
                sample_inference_image_path = valid_images[0]
                print(f"Usando imagen para inferencia: {sample_inference_image_path}")

                results_inference = trained_model(sample_inference_image_path, max_det=10)

                for r in results_inference:
                    im_array = r.plot()
                    plt.figure(figsize=(10,10))
                    plt.imshow(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB))
                    plt.title(f"Detecciones en {os.path.basename(sample_inference_image_path)}")
                    plt.axis('off')
                    plt.show()
            else:
                print(f"No se encontraron imÃ¡genes en el directorio de validaciÃ³n para inferencia: {valid_images_path}")
        else:
            print(f"Directorio de imÃ¡genes de validaciÃ³n no encontrado: {valid_images_path}")
    else:
        print(f"El modelo entrenado ({BEST_MODEL_PATH if BEST_MODEL_PATH else 'ruta no encontrada'}) no fue encontrado. Â¿El entrenamiento se completÃ³ correctamente?")

except Exception as e:
    print(f"Error en el paso 5 (evaluaciÃ³n/inferencia): {e}")
    import traceback
    traceback.print_exc()

print("\n--- Proceso Finalizado ---")
```

```python
# ==============================================================================
# PASO 0: INSTALACIÃ“N DE LIBRERÃAS NECESARIAS
# ==============================================================================
print("Paso 0: Instalando librerÃ­as...")
!pip install ultralytics -q # -q para modo silencioso
!pip install ffmpeg-python -q # Para asegurar que ffmpeg estÃ¡ bien configurado para OpenCV y YOLO

import os
import cv2
from ultralytics import YOLO
from IPython.display import HTML, display
from base64 import b64encode
import shutil # Para limpiar directorios si es necesario

print("LibrerÃ­as instaladas y mÃ³dulos importados.")

# ==============================================================================
# PASO 1: CONFIGURACIÃ“N DE RUTAS
# ==============================================================================
print("\nPaso 1: Configurando rutas...")

# --------- MODIFICA ESTAS RUTAS SEGÃšN SEA NECESARIO ---------
# Ruta a tu modelo .pt entrenado (best.pt)
# Sube tu archivo best.pt a Colab y ajusta esta ruta.
# Ejemplo: Si lo subes a la raÃ­z de /content/
MODEL_PATH = '/content/best.pt'  # <<<<<<< Â¡ASEGÃšRATE DE QUE ESTA RUTA SEA CORRECTA!

# Ruta a tu video de entrada
# Sube tu video .mov a Colab y ajusta esta ruta.
# Ejemplo: Si lo subes a la raÃ­z de /content/
VIDEO_INPUT_PATH = '/content/GH012372.mp4' # <<<<<<< Â¡REEMPLAZA ESTO CON LA RUTA A TU VIDEO!

# Directorio donde se guardarÃ¡n los resultados del procesamiento de video
# Los videos procesados se guardarÃ¡n en subdirectorios dentro de esta carpeta.
OUTPUT_PROJECT_DIR = "/content/runs/detect_video"
OUTPUT_RUN_NAME = "video_processed_output" # Nombre de la subcarpeta especÃ­fica para esta ejecuciÃ³n
# -----------------------------------------------------------

# Limpiar directorio de salida anterior si existe (opcional, para evitar video_output2, video_output3...)
# output_specific_dir = os.path.join(OUTPUT_PROJECT_DIR, OUTPUT_RUN_NAME)
# if os.path.exists(output_specific_dir):
#     print(f"Limpiando directorio de salida anterior: {output_specific_dir}")
#     shutil.rmtree(output_specific_dir) # Â¡Cuidado! Esto borra la carpeta y su contenido.

os.makedirs(OUTPUT_PROJECT_DIR, exist_ok=True)

print(f"Ruta del modelo: {MODEL_PATH}")
print(f"Ruta del video de entrada: {VIDEO_INPUT_PATH}")
print(f"Los resultados se guardarÃ¡n en subcarpetas de: {OUTPUT_PROJECT_DIR}")

# ==============================================================================
# PASO 2: CARGAR EL MODELO ENTRENADO
# ==============================================================================
print("\nPaso 2: Cargando el modelo entrenado...")

if not os.path.exists(MODEL_PATH):
    print(f"Error: El archivo del modelo no se encuentra en: {MODEL_PATH}")
    print("Por favor, sube tu archivo 'best.pt' (o el nombre de tu modelo) a Colab y verifica la ruta.")
    raise SystemExit("Modelo no encontrado.")

try:
    model = YOLO(MODEL_PATH)
    print(f"Modelo cargado exitosamente desde: {MODEL_PATH}")
except Exception as e:
    print(f"Error crÃ­tico al cargar el modelo desde {MODEL_PATH}: {e}")
    raise SystemExit("Fallo al cargar el modelo.")

# ==============================================================================
# PASO 3: VERIFICAR EL VIDEO DE ENTRADA
# ==============================================================================
print("\nPaso 3: Verificando el video de entrada...")

if not os.path.exists(VIDEO_INPUT_PATH):
    print(f"Error: El archivo de video no se encuentra en: {VIDEO_INPUT_PATH}")
    print("Por favor, sube tu video a Colab y verifica la ruta.")
    raise SystemExit("Video de entrada no encontrado.")
else:
    print(f"Video de entrada encontrado: {VIDEO_INPUT_PATH}")

    # Intenta abrir el video con OpenCV para una verificaciÃ³n preliminar
    cap = cv2.VideoCapture(VIDEO_INPUT_PATH)
    if not cap.isOpened():
        print(f"ADVERTENCIA: OpenCV no pudo abrir el video directamente en: {VIDEO_INPUT_PATH}")
        print("Esto podrÃ­a indicar un problema con el formato del video o las dependencias.")
        print("YOLO podrÃ­a intentar procesarlo igualmente usando sus propios manejadores.")
    else:
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        print(f"  OpenCV pudo abrir el video: {width}x{height} @ {fps:.2f} FPS, {frame_count} frames")
    cap.release()

# ==============================================================================
# PASO 4: PROCESAR EL VIDEO CON YOLO
# ==============================================================================
print("\nPaso 4: Iniciando procesamiento del video con YOLO...")

try:
    # stream=True es eficiente para videos largos, procesa frame por frame
    # save=True guardarÃ¡ el video con las detecciones
    # project: directorio base para los resultados
    # name: nombre de la subcarpeta especÃ­fica para esta ejecuciÃ³n
    # exist_ok=True: si la carpeta 'name' ya existe, la usarÃ¡ y sobrescribirÃ¡.
    #                Si es False (default), crearÃ¡ name2, name3, etc.
    # conf: umbral de confianza para las detecciones (ej. 0.25 significa >25% de confianza)
    # iou: umbral de IoU para Non-Maximum Suppression (NMS)
    results_generator = model.predict(
        source=VIDEO_INPUT_PATH,
        stream=True,
        save=True,
        project=OUTPUT_PROJECT_DIR,
        name=OUTPUT_RUN_NAME,
        exist_ok=True, # Cambia a False si prefieres que cree carpetas nuevas (video_output2, etc.)
        conf=0.3,      # Ajusta el umbral de confianza segÃºn necesites
        iou=0.5        # Ajusta el umbral de IoU segÃºn necesites
    )

    print(f"Procesando frames del video '{os.path.basename(VIDEO_INPUT_PATH)}'...")
    # Es necesario iterar sobre el generador para que el procesamiento y guardado ocurran.
    processed_frames_count = 0
    for i, result in enumerate(results_generator):
        processed_frames_count += 1
        if (i + 1) % 100 == 0: # Imprime un progreso cada 100 frames
            print(f"  Procesado frame {i+1}...")
        # result contiene informaciÃ³n sobre las detecciones en el frame actual.
        # Puedes acceder a result.boxes, result.masks, result.probs si necesitas
        # informaciÃ³n detallada por frame.
        # Por ahora, solo estamos interesados en el video guardado.
    print(f"Procesamiento de video completado. Total frames procesados: {processed_frames_count}")

except Exception as e:
    print(f"OcurriÃ³ un error durante el procesamiento del video con YOLO: {e}")
    import traceback
    traceback.print_exc()
    raise SystemExit("Fallo durante el procesamiento del video.")

# ==============================================================================
# PASO 5: ENCONTRAR Y MOSTRAR EL VIDEO PROCESADO
# ==============================================================================
print("\nPaso 5: Mostrando el video procesado...")

# La ruta esperada del video de salida
# YOLOv8 guarda el video con el mismo nombre que el original dentro de la carpeta de 'name'
expected_output_filename = os.path.basename(VIDEO_INPUT_PATH)
video_output_path = os.path.join(OUTPUT_PROJECT_DIR, OUTPUT_RUN_NAME, expected_output_filename)

if os.path.exists(video_output_path):
    print(f"Video procesado encontrado en: {video_output_path}")
    try:
        mp4 = open(video_output_path, 'rb').read()
        data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

        display(HTML(f"""
        <div style="text-align: center;">
            <h3>Video Procesado: {os.path.basename(video_output_path)}</h3>
            <video width="80%" max-width="720px" controls style="border:1px solid #ccc;">
                  <source src="{data_url}" type="video/mp4">
                  Tu navegador no soporta el elemento de video.
            </video>
            <p>Puedes descargar el video desde el explorador de archivos en Colab, en la ruta:</p>
            <p><code>{video_output_path}</code></p>
        </div>
        """))
    except Exception as e:
        print(f"Error al intentar mostrar el video en Colab: {e}")
        print(f"Puedes descargar el video manualmente desde el explorador de archivos en: {video_output_path}")
else:
    print(f"Error: No se pudo encontrar el video de salida procesado en la ruta esperada: {video_output_path}")
    print("Verifica la consola por errores durante el procesamiento.")
    # Listar contenido de la carpeta de salida para depuraciÃ³n
    output_run_folder = os.path.join(OUTPUT_PROJECT_DIR, OUTPUT_RUN_NAME)
    if os.path.exists(output_run_folder):
        print(f"Contenido de la carpeta de salida '{output_run_folder}': {os.listdir(output_run_folder)}")
    else:
        print(f"La carpeta de salida '{output_run_folder}' no existe.")

print("\n--- Proceso de inferencia de video Finalizado ---")
```
